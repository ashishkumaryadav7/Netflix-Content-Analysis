# Data Analysis & Visualization

#### How many of the total TV shows & Movies availabe

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(12,6))
total_tm = df['type'].value_counts()
total_tm.plot(kind='bar',width=.4,color=['#bf5770','#caa0cd'])
plt.title("Total TV shows & Movies")
plt.xlabel("Type")
plt.ylabel("No. of TV shows & Movies")

# total_tm.plot.pie(shadow = True, autopct='%1.2f%%')
plt.show()
```

#### which year highest number of the TV shows & movies release only top 10 records

```python
df.columns
```

```python
sort_index = df['release_year'].value_counts().index.sort_values(ascending=False)
year_count = df['release_year'].value_counts()
```

```python
sort_index_series = pd.Series(sort_index)
sort_index_series_top10 = sort_index_series.head(10)
year_count_top10 = year_count.loc[sort_index_series_top10]
```

```python
print(year_count_top10)
print(sort_index_series_top10)
```

```python
sns.set_style(style="dark")
plt.figure(figsize=(12,6))
ax1 = sns.barplot(x=sort_index_series_top10 , y= year_count_top10.values, order=sort_index_series_top10)
for container in ax1.containers:
    ax1.bar_label(container)
plt.title("Most released TV shows & Movies in last 10 years")
plt.xlabel("Release Years")
plt.ylabel("No. of TV shows & movies")
plt.show()
```

## Rating of TV shows and movies 

### 1. Top 5 Rating of TV shows that are given to TV shows

```python
TVshow_dur_df['rating'].unique()
TVshow_dur_df.listed_in.head()
```

```python
# removing the null value in 'rating' column
rating_TVshow_df = TVshow_dur_df.dropna(subset=['rating'])
```

```python
plt.figure(figsize=(8,8))
plt.pie(data=rating_TVshow_df,
        x=rating_TVshow_df['rating'].value_counts().head(),
        autopct='%0.2f%%',startangle=30,
        labels = (rating_TVshow_df['rating'].value_counts().head()).index,
        rotatelabels = True,
        shadow=True,
       explode= [0,0,0,0,0])
plt.title("Maximum TV-Shows that contain rating (Top 5 rating)",fontdict={'color':"y"})
plt.legend()
plt.show()
```

```python
def sep(row):
    if "Comedies" in row['listed_in']:
        return ("Comedies")
    elif "Dramas" in row['listed_in']:
        return ("Dramas")
    elif "Crime TV Shows" in row['listed_in']:
        return ('Crime TV Shows')
    elif 'Horror' in row['listed_in']:
        return ("Horror")
    elif "Kids" in row['listed_in']:
        return ("Kids")
    elif "Sci-Fi" in row['listed_in']:
        return ("Sci-Fi")
    elif 'TV Action & Adventure' in row['listed_in']:
        return ("TV Action & Adventure")
    else:
        return ("other")
```

```python
rating_TVshow_df['new_c'] = rating_TVshow_df.apply(sep,axis=1)

# group by column 'rating' & 'new_c'
grouped = rating_TVshow_df.groupby(['rating', 'new_c']).size()
```

```python
#Dramas,#Children & Family Movies,International Movies,Horror Movies,Action & Adventure
def sep1(row):
    if "Comedies" in row['listed_in']:
        return ("Comedies")
    elif "Dramas" in row['listed_in']:
        return ("Dramas")
    elif "Children & Family Movies" in row['listed_in']:
        return ('kids & Family Movies')
    elif 'Horror' in row['listed_in']:
        return ("Horror")
    elif "International Movies" in row['listed_in']:
        return ("International")
    elif "Sci-Fi" in row['listed_in']:
        return ("Sci-Fi")
    elif 'Action & Adventure' in row['listed_in']:
        return ("Action & Adventure")
    else:
        return ("other")
```
